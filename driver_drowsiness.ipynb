```python
import cv2
import dlib
import imutils
from pygame import mixer
from imutils import face_utils
from scipy.spatial import distance
```

```python
mixer.init()
mixer.music.load('music.wav')
```

```python
def calculate_EAR(eye):
	A = distance.euclidean(eye[1], eye[5])
	B = distance.euclidean(eye[2], eye[4])
	C = distance.euclidean(eye[0], eye[3])
	ear = (A+B)/(2.0*C)
	return ear
```

```python
modelpath="shape_predictor_68_face_landmarks (1).dat"
```

```python
hog_face_detector=dlib.get_frontal_face_detector()
dlib_facelandmark=dlib.shape_predictor(modelpath)
```

```python
cap=cv2.VideoCapture(0)
while True:
    ret,frame=cap.read()
    if ret== False:
        break
    
    
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=hog_face_detector(gray)
    for face in faces:
        face_landmarks=dlib_facelandmark(gray, face)
        left_eye=[]
        right_eye=[]
        for n in range(36,42):
            x=face_landmarks.part(n).x
            y=face_landmarks.part(n).y
            left_eye.append((x,y))
            next_point=n+1
            if n==41:
                next_point=36
           
            x1=face_landmarks.part(next_point).x
            y1=face_landmarks.part(next_point).y
            cv2.line(frame,(x,y),(x1,y1),(0,255,0),1)
        for n in range(42,48):
            x=face_landmarks.part(n).x
            y=face_landmarks.part(n).y
            right_eye.append((x,y))
            next_point=n+1
            if n==47:
                next_point=42
           
            x2=face_landmarks.part(next_point).x
            y2=face_landmarks.part(next_point).y
            cv2.line(frame,(x,y),(x2,y2),(0,255,0),1)
        left_ear= calculate_EAR(left_eye)
        right_ear=calculate_EAR(right_eye)
        avg=(left_ear+right_ear)/2
        thresh=0.26
       
        
        if avg< thresh:
            cv2.putText(frame, "****************ALERT!****************", (10, 30),cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
            mixer.music.play()
			
			
            
        print(avg)
    
        
        
       
    cv2.imshow("are you sleepy",frame)
    if cv2.waitKey(1)==ord('q'):
        break
    
cap.release()
cv2.destroyAllWindows
```
